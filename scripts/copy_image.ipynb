## Copy images

```python
def copiar_imagens(source_folder, dest_folder, image_list, subset):

  print(f"len(image_list): {len(image_list)}")
  attributes = ["_attribute_globules", "_attribute_milia_like_cyst", "_attribute_negative_network", "_attribute_pigment_network", "_attribute_streaks"]

  if subset == "train":
    source_path_image = f"{source_folder}/ISIC2018_Task1-2_Training_Input"
    source_path_mask = f"{source_folder}/ISIC2018_Task2_Training_GroundTruth_v3"

    dest_path_image = f"{dest_folder}/ISIC2018_Task1-2_Training_Input"
    dest_path_mask = f"{dest_folder}/ISIC2018_Task2_Training_GroundTruth_v3"
  elif subset == "val":
    source_path_image = f"{source_folder}/ISIC2018_Task1-2_Validation_Input"
    source_path_mask = f"{source_folder}/ISIC2018_Task2_Validation_GroundTruth"

    dest_path_image = f"{dest_folder}/ISIC2018_Task1-2_Validation_Input"
    dest_path_mask = f"{dest_folder}/ISIC2018_Task2_Validation_GroundTruth"

  if not os.path.exists(dest_path_image):
    os.makedirs(dest_path_image)
  if not os.path.exists(dest_path_mask):
    os.makedirs(dest_path_mask)

  for image_file in image_list:
    source_file_path_image = os.path.join(source_path_image, image_file)
    dest_file_path_image = os.path.join(dest_path_image, image_file)

    if os.path.exists(source_file_path_image):
      shutil.copy2(source_file_path_image, dest_file_path_image)
      print(f'Imagem "{image_file}" copiada com sucesso para a pasta de destino.')

    for attribute in attributes:
      image_file_aux = image_file.split(".")[0] + attribute + ".png"
      source_file_path_mask = os.path.join(source_path_mask, image_file_aux)
      dest_file_path_mask = os.path.join(dest_path_mask, image_file_aux)

      if os.path.exists(source_file_path_mask):
        shutil.copy2(source_file_path_mask, dest_file_path_mask)
        print(f'Mascara "{image_file_aux}" copiada com sucesso para a pasta de destino.')
```

```python
def elementos_em_pelo_menos_duas_listas(*listas):
    if len(listas) < 2:
        return []

    conjunto_contagem = {}

    for lista in listas:
        for elemento in lista:
            conjunto_contagem[elemento] = conjunto_contagem.get(elemento, 0) + 1

    elementos_repetidos = [elemento for elemento, count in conjunto_contagem.items() if count >= 2]
    return elementos_repetidos

elementos_rep = elementos_em_pelo_menos_duas_listas(training_attribute_globules, training_attribute_milia_like_cyst, training_attribute_negative_network, training_attribute_pigment_network, training_attribute_streaks)
print("Elementos presentes em pelo menos duas listas:", len(elementos_rep))
```
