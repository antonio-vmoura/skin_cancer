## Counts attributes of masks

```python
def count_attributes(img_ids, data_masks, flag="_segmentation"):
    occurrence_list = []

    for n, ids in enumerate(tqdm(img_ids)):
        mask = load_img(os.path.join(data_masks, ids.split(".")[0] + flag + ".png")) #load the mask
        clrs = mask.getcolors()

        if len(clrs) != 1: #if there is more than one color it is because it occurs
          occurrence_list.append(ids)

    return occurrence_list
```
